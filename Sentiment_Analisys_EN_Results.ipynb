```python
lemmatizer = WordNetLemmatizer()

for i in range(new_size):
  try:
    video = new_videoid.iloc[i]
    name = video +'.csv'
    dt = pd.read_csv(name)
    processed_word = list()
    for op in dt.text:
      new_text = list()
      #words_text = token_space.tokenize(op.translate (str.maketrans ('', '', string.punctuation)))
      #Pontuação
      #falta minusculos
      words_text = token_space.tokenize(unidecode(op).lower())
      for words in words_text:
        if words not in non_words:
          new_text.append(words)
      #singles = [stemmer.stem(plural) for plural in new_text]
      lemmatized_words = [lemmatizer.lemmatize(word) for word in new_text]

      processed_word.append(' '.join(lemmatized_words))


    dt["treatment_1"] = processed_word
    dt.to_csv(name, index=False)
  except:
    print('erro')
    i+=1
```

```python
sid = si()
for i in range(new_size):
  try:
    video = new_videoid.iloc[i]
    name = video +'.csv'
    dt = pd.read_csv(name)
    score_word = list()
    sentance = dt.loc[:,"treatment_1"]
    t = sentance.size
    for k in range(t):
      word = sentance.iloc[k]
      score = sid.polarity_scores(str(word))
      score_word.append(score)

    dt["Score"] = score_word
    dt.to_csv(name, index=False)
  except:
    i+=1
```

```python
for i,row in df_new.iterrows():
  try:
    video_id = row['ID']
    video_request = youtube.videos().list(
          part='snippet,statistics',
          id=video_id
    )

    video_response = video_request.execute()
    likes = video_response['items'][0]['statistics']['likeCount']
    views = video_response['items'][0]['statistics']['viewCount']

    df_new.loc[i, 'Likes'] = int(likes)
    df_new.loc[i, 'Views'] = int(views)
  except:
    df_new.loc[i, 'Likes'] = 0
    df_new.loc[i, 'Views'] = 0
    i+= 1
```

```python
disabled_comments_indices = []

for i,row in df_new.iterrows():

  video_id = row['ID']
  try:
      # Fetch the comments
      comments_request = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=3000,  # Limit to 3000 comments
        textFormat='plainText'
        )
      comments_response = comments_request.execute()

        # Extract the comments
      comments = [item['snippet']['topLevelComment']['snippet']['textDisplay']
                    for item in comments_response['items']]

        # Create a DataFrame for the comments
      comments_df = pd.DataFrame(comments, columns=['Comment'])

        # Save the comments to a CSV file
      comments_df.to_csv(f'{video_id}_comments.csv', index=False)

  except HttpError as e:
      if 'commentsDisabled' in str(e):
          # If comments are disabled for the video, add its index to the list
          disabled_comments_indices.append(i)

# Remove videos with disabled comments from the original DataFrame
df_new.drop(disabled_comments_indices, inplace=True)

# Save the updated DataFrame to a new CSV file
df_new.to_csv('updated_file.csv', index=False)
```
