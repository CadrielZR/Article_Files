```python
import pandas as pd
```

```python
dataset=pd.read_csv('df_new_scores_pt (1).csv', engine='c', sep=',')
dataset
```

```python
dataset.corr(method='spearman')
```

```python
#https://www.youtube.com/watch?v=njt6wKWuCp8&ab_channel=DataScienceGarage
#https://geographyfieldwork.com/SpearmansRankCalculator.html#:~:text=The%20p%20(or%20probability)%20value,null%20hypothesis%20assumption%20is%20correct.
import scipy
df = pd.DataFrame()
feat1s = []
feat2s = []
corrs = []
p_values = []

for feat1 in dataset.columns:
    for feat2 in dataset.columns:
        if feat1 != feat2:
            feat1s.append(feat1)
            feat2s.append(feat2)
            corr, p_value = scipy.stats.spearmanr(dataset[feat1], dataset[feat2])
            corrs.append(corr)
            p_values.append(p_value)

df['Feature_1'] = feat1s
df['Feature_2'] = feat2s
df['Correlation'] = corrs
df['p_value'] = p_values
df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize']= 18,5
plt.plot(df.index, df['p_value'], linewidth=0, marker='D', markersize=10)
plt.grid(which='major', color='#cccccc', linestyle='--')
plt.title('Correlations with calculated p-value')
plt.xlabel('Feature pairs')
plt.ylabel('p-value')
plt.xticks(df.index)
plt.show()
```

```python
p_values=df['p_value'].values
Feature_1=df['Feature_1'].values
Feature_2=df['Feature_2'].values
Correlation=df['Correlation'].values

for i in (df.index):
  if (p_values[i]<0.01):
    if (Feature_1[i] in ['Neutral Score', 'Positive Score', 'Negative Score', 'Compund Score']):
      if (Feature_2[i] in ['Like/Views',	'Comment Score Neg', 'Comment Score Pos',	'Comment Score Neu',	'Comment Score Compund']):
        print(i,'/', Feature_1[i], '/', Feature_2[i], Correlation[i], '/', p_values[i])
        print('Rejeitar H0 = Pois existe correlação')
```
